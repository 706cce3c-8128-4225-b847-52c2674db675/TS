## Определение $\Delta$ модели HIT по данным заболеваемости в регионах Росcии
В данном ноутбуке изучается возможность применения $HIT$ модели из статьи Юрия Нестерова (https://arxiv.org/abs/2007.11429), а именно: делается попытка установить значения константы $\Delta$ из статистики заболеваемости в регионах России.

Метод:  
Мы имеем статистику по кол-ву новых случаев заражения для разных городов. По этой статистике мы можем посчитать $\gamma$ при установленном $\Delta$. Если мы предполагаем, что в какой-то день $\gamma$ должна сильно изменится (причем эти изменения не являются однодневным скачком, а являются изменением в поведении общества, то есть их эффект будет держаться некоторое количество дней), то мы можем посчитать среднее (или взвешенное среднее) для $\gamma$ с разных сторон от события. И потом выбрать дельта исходя из того, где эта разность будет наиболее "большой". "Большой" написано в кавычках, так как нельзя непосредственно сравнить разности между собой, абсолютные значения $\gamma$ сильно зависят от выбранного $\Delta$.

Мы сделаем предположение, что хоть абсолютные изменения $\gamma$ и нельзя сравнивать, но можно сравнивать относительные изменения: среднее_справа - среднее_слева / (среднее_справа + среднее_слева). Это предположение может и не выполняться, хотя кажется более-менее правдоподобным.   

Вышеописанным образом определяется дельта для разных регионов, опираясь на разные даты: 
* день единого голосования, 
* день начала нерабочей недели
* первое сентября.  

Подсчитывается, сколько раз встретилось каждое значение $\Delta$, и тем самым получается некоторое распределение, которое. В случае, если все сделанне предположение верны с некоторой точностью, это распределение будет иметь явно выраженную моду, которую можно было бы считать оценкой $\Delta$.

При просмотре графиков таже было замечено, что не для всех регионов и предполагаемых дат изменения $\gamma$, это изменение можно наблюдать. Чтобы убрать из подсчета такие случаи предлагается следующий способ:   
Берутся множества:

$\Gamma_{left} = \{\gamma(d-n), ..., \gamma(d-1)\}$  
$\Gamma_{right} = \{\gamma(d), ..., \gamma(d+n)\}$  
$\Gamma_{full} = \{\gamma(d-n), ..., \gamma(d+n)\}$  
где в $\gamma(d)$ - значение $\gamma$ в $d$-ый день с начала ведения статистики.  
 
Cравнивается дисперсия множества $\Gamma_{full} - \overline{\Gamma_{full}}$ с множеством $(\Gamma_{left} - \overline{\Gamma_{left}}) \cup (\Gamma_{right} - \overline{\Gamma_{right}})$  
где черта сверху означает выборочное среднее. 

Сравнение дисперсий производится с использованием критерия Левина. Если $p-value$ критерия достаточно маленький, то мы отвергаем гипотезу о равенстве дисперсий и рассматриваем такой случай.

Так же были высказаны предположения, что в городах с маленьким населением могут быть сильные колебания $\gamma$, так как там несколько новых случаев уже является значимым явлением. И, возможно, их также необходимо исключить из рассмотрения.

Так же была высказана идея, что лучше брать не одно $\Delta$, при котором достигается самое большое значение разности, а top k таких $\Delta$ и брать статистику уже по ним.

````python
def minmax_scale(arr, scale = 1.):
    tmp = arr - np.min(arr)
    return scale * tmp / np.max(tmp)
````

## Подбор коэффициентов сглаживания

````python
def side_smooth(arr, side, size = 6, lam = 1.0, kernel='uniform'):
    
        assert size > 0, 'size должен быть больше нуля!'
        
        if len(arr) == 1:
            return arr[0]
        
        if size == 1:
            return arr[0] if side == 'right' else arr[-1]
        
        size = min(size, len(arr))
        
        if kernel == 'uniform':
            coefs = np.ones((size,))
            coefs /= np.sum(coefs)
        elif kernel == 'poisson':
            coefs = np.array([np.exp(-lam)*(lam**k)/np.math.factorial(k) for k in range(size)])
            coefs /= np.sum(coefs)
        elif kernel == 'norm':
            coefs = norm.cdf(np.linspace(0, -3, size+1))
            coefs = coefs[1:] - coefs[:-1]
            coefs /= np.sum(coefs)
        else:
            raise NameError('Непонятное ядро.')
            
        if side == 'left':
            val = np.sum(arr[-size:]*coefs[::-1])
        elif side == 'right':
            val = np.sum(arr[:size]*coefs)
        else:
            raise NameError("side может принимать только значения 'left' и 'right'")
        
        assert np.max(coefs[1:] -coefs[:-1]) < 10e-7, "коэффициенты возрастают при удалении от целевого дня!"
        return val
````

````python
def plot_jump_from_delta(time_point,
                        reg = 'Москва',
                        win_s =10,
                        delta = 7,
                        smooth_size = 12,
                        smooth_lam = 0.95,
                        smooth_kernel = 'poisson'):
    
    fig, ax = plt.subplots()
    fig.set_size_inches(16, 9)
    ax.grid(True)
    ax.grid(True,'minor')

    #Берем известные данные 
    d, T = Globals.get_covid_region_data(reg)[['date', 'cases_total']].values.T
    
    #Считаем гамма с учетом заданного delta
    H =  HIT(d, T, delta)
    d, gammas = H.get_gamma()

    #Устраняем nan
    gammas = np.nan_to_num(gammas, 0)


    #Находим нужный день
    print(time_point)
    tp_id = np.where(d == pd.to_datetime(time_point))[0][0]
    

    #Считаем значение сглаженного графика
    left_s = max(tp_id-win_s-1, 0)
    smooth_gammas = [side_smooth(gammas[:i+1], 'left',
                          size=smooth_size,
                          lam = smooth_lam,
                          kernel= smooth_kernel) for i in range(left_s, tp_id)]

    smooth_gammas += [side_smooth(gammas[i:], 'right',
                          size=smooth_size,
                          lam = smooth_lam,
                          kernel= smooth_kernel) for i in range(tp_id, min(tp_id+win_s, len(gammas)))]
    
    smooth_gammas = np.array(smooth_gammas)


    ax.plot(d[left_s+2:tp_id+win_s], gammas[left_s+2:tp_id+win_s],
                                         label = 'истинные значения (delta = {0})'.format(delta))
    ax.plot(d[left_s+2:tp_id+win_s], smooth_gammas[2:], 
                                         label = 'сглаженные значения (delta = {0})'.format(delta))
    
    ax1 = ax.twinx()
    ax1.plot(d[left_s+2:tp_id+win_s], minmax_scale(smooth_gammas[2:] - smooth_gammas[1:-1], scale = 1),
                    linestyle = '--',  color='b', linewidth=1, label = 'разность с предыдущим днем')


    ax1.plot([time_point, time_point], [0, 1], label = 'Предполагаемый день изменения гамма',
                                               linestyle = '--',  color='r', linewidth=2)

    ax.legend(loc=2)
    ax1.legend(loc=0)
    plt.show()
````

````python
graph_func = lambda l, k, ss, d, r: plot_jump_from_delta(reg = r,
                            win_s =15,
                            delta = l,
                            time_point = interesting_dates[d][0],
                            smooth_size = ss,
                            smooth_lam = 0.95,
                            smooth_kernel = k)
````

````python
#Проверка возможности применения теста Левина (проверка равенства знаменятеля статистики нулю)
def levene_denom_zero(args):
    k=2
    Ni = np.zeros(k)
    Yci = np.zeros(k)
    func = lambda x: np.mean(x, axis=0)

    for j in range(k):
        Ni[j] = len(args[j])
        Yci[j] = func(args[j])
    Ntot = np.sum(Ni, axis=0)

    # compute Zij's
    Zij = [None] * k
    for i in range(k):
        Zij[i] = abs(np.asarray(args[i]) - Yci[i])

    # compute Zbari
    Zbari = np.zeros(k, 'd')
    Zbar = 0.0
    for i in range(k):
        Zbari[i] = np.mean(Zij[i], axis=0)
        Zbar += Zbari[i] * Ni[i]

    Zbar /= Ntot
    numer = (Ntot - k) * np.sum(Ni * (Zbari - Zbar)**2, axis=0)

    # compute denom_variance
    dvar = 0.0
    for i in range(k):
        dvar += np.sum((Zij[i] - Zbari[i])**2, axis=0)

    return  abs(dvar) < 10e-5
````

````python
def define_delta(min_d=1, max_d=20, min_population=10e7,
                 lam=0.95,
                 smooth_size=4,
                 kernel='uniform',
                 top=1,
                 p_val=0.05,
                 print_table=True,
                 draw_distr=True,
                 draw_pval_distibution=False):
    if print_table:
        table = '''Регион | Дата  | Топ дельт с максимальным скачком среднего | Население\n'''

    reg_deltas = list([])
    lev_list = list([])

    total, total_ignored = 0, 0
    ignored = {i: 0 for i in interesting_dates}

    # По все регоинам
    for reg in Globals.region_friendly_name_inv:

        if min_population > Globals.region_population[reg]:
            continue

        dT, T = Globals.get_covid_region_data(reg)[['date', 'cases_total']].values.T

        # По всем событиям, при которых предполагается изменение gamma
        # --------------------------------------------------------------------------------
        for day_of_interest in interesting_dates:

            time_point, tendency = interesting_dates[day_of_interest]

            # -----------------определение, есть ли скачек----------------
            tp_id = np.where(dT == time_point)[0][0]
            n = smooth_size
            left_s, right_s = max(tp_id - n, 0), min(tp_id + n, len(T))
            left, right, full = T[left_s:tp_id].copy(), T[tp_id:right_s].copy(), T[left_s: right_s].copy()

            left = left - np.mean(left)
            right = right - np.mean(right)
            full = full - np.mean(full)
            merged = np.hstack([left, right])

            total += 1

            # Если p_val - достаточно высокий, то
            # то мы не можем отвергнуть гипотизу об
            # одинаковых дисперсиях

            try:
                exclude = levene(full, merged)[1] > p_val
            except:
                exclude = True
                   
            if exclude:
                ignored[day_of_interest] += 1
                total_ignored += 1
                continue

            lev_list += [levene(full, merged)[1], ] * top
            # ---------------------------------------------------

            jumps = list([])
            
            #Оценка скачка при разных Delta
            for delta in range(min_d, max_d + 1):

                d, gammas = HIT(dT, T, delta).get_gamma()

                gammas = np.array(gammas, dtype=np.float32)
                gammas = np.nan_to_num(gammas, 0)

                try:
                    tp_id = np.where(d == pd.to_datetime(time_point))[0][0]
                except:
                    # Выскакивает, когда нет информации за нужную дату.
                    if np.sum(d == time_point) == 0:
                        print('Для региона {0} и даты {1} нет данных в статистике'.format(reg, time_point))
                    continue
                
                r = side_smooth(gammas[tp_id:], 'right',
                                size=smooth_size, lam=lam,
                                kernel=kernel)

                l = side_smooth(gammas[:tp_id], 'left',
                                size=smooth_size, lam=lam,
                                kernel=kernel)
                
                
                if (abs(r) + abs(l)) == 0: #Такое бывает, для малонаселенных регионов.
                    jump = 0
                else:
                    jump = 2 * (r - l) / (abs(r) + abs(l))
 
                jumps.append(jump)
            
            # ----------------------------------------------------------------------------------------

            jumps = np.array(jumps)

            if tendency == 'up':
                reg_deltas += list(np.argsort(jumps)[-top:] + min_d)
            elif tendency == 'down':
                reg_deltas += list(np.argsort(-jumps)[-top:] + min_d)
            else:
                raise NameError('Неизвесная тенденция : {0}'.format(tendency))

            if print_table:
                table = table + '''{0} | {1} | {2} | {3}\n'''.format(reg,
                                                                     time_point, np.argsort(jumps)[-top:] + min_d,
                                                                     Globals.region_population[reg])

    for day in ignored:
        print('Отсеяно для  {0} : {1}'.format(day, ignored[day]))
    print('\n Отсеяно тестом Левина : {0} \ Общее кол-во кейсов: {1}'.format(total_ignored, total))

    
    #Отрисовка графиков.
    
    #Распределение Delta
    if draw_distr:
        vals, counts = np.unique(reg_deltas, return_counts=True)
        plt.figure(figsize=(10, 7))
        plt.plot(vals, counts, label='num of delta')
        plt.ylabel('кол-во delta полученное для разных регионов и дат')
        plt.xlabel('значение delta')
        plt.hist(reg_deltas, label='hist')
        plt.legend()
        plt.show()

    #Распределение p-value в зависимости от Delta
    if draw_pval_distibution:
        x = {i: list([]) for i in np.unique(reg_deltas)}
        for i in range(len(reg_deltas)):
            x[reg_deltas[i]].append(lef_list[i])

        tx, y, y_max, y_min = list([]), list([]), list([]), list([])
        for i in x:
            tx.append(i)
            y.append(np.nanmean(x[i]))
            y_max.append(np.nanmax(x[i]))
            y_min.append(np.nanmin(x[i]))
            plt.plot(tx, y)
            plt.plot(tx, y_max)
            plt.plot(tx, y_min)
        plt.show()

    if print_table:
        print(table)
````

````python
delta_func = lambda k, ss, dmin, dmax, p_val, t, tab, p : define_delta(min_d=dmin, max_d=dmax,
                                                                 min_population = p,
                                                                 lam=1.0,
                                                                 smooth_size=ss,
                                                                 kernel= k,
                                                                 top=t,
                                                                 p_val = p_val,
                                                                 print_table=tab,
                                                                 draw_distr=True,
                                                                 draw_pval_distibution = False)
````

2. Та мода, что получается при некоторых значениях параметров, очень сильно зависит от самих параметров.
